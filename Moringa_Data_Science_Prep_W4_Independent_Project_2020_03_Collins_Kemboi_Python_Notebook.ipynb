# **1. Validity**

```
# Dropping the Display comment, subscribtion status, status and Scheduled at columns : We will not need the mentioned columns
# anywhere for our analysis.
# 
# 
df.drop(['Displayed comment', 'Scheduled at', 'Status','Subscription status' ], axis = 1, inplace = True)


df
```

# **3. Completeness**

```
# Data Cleaning Action: Check for duplicate values.
# Explanation: This is to allow us determine data that is duplicated.
#
df_duplicates = df.duplicated().sum()
df_duplicates
```

# **4. Consistency**

```
# Data Cleaning Action: Dropping null values if they exist.

#
df_new = df.dropna(how="all")
df_new.head()
```

# **5. Uniformity**

```
# Data Cleaning Action: Converting column names to lowercase characters.
# Explanation: This is to allow for a uniform columns in the dataframe.
#
df_new.columns = df_new.columns.str.rstrip().str.lower()
df_new.head(5)
```

```
# Data Cleaning Action: Converting row values to lowercase characters.
# Explanation: This is to allow for rows to have uniform values which are in lowercase in the dataframe.
#
df_new = df_new.astype(str).apply(lambda x: x.str.lower())
df_new.head(5)
```

# Exporting our clean csv file

```
df_new.to_csv("autolib_clean.csv")
```

# Performing Analysis

```
url = "autolib_clean.csv"
data = pd.read_csv(url)
data.head()
```

```
#Identify the most popular hour of the day for picking up a shared electric car 
# (Bluecar) in the city of Paris over the month of April 2018.*

df_p = data[['city', 'bluecar counter', 'hour']]
df_paris = data[data['city'] == 'paris']
df_p = df_paris.groupby(['bluecar counter'])['hour'].max().sort_values(ascending=False)
print(df_p)
```

```
# What is the most popular hour for returning cars?


df_most = data[["bluecar counter","day", "hour", "minute"]]
df_car = df_most[df_most["bluecar counter"] == 0]
df_car['hour'].max()
```

```
# Finding the most popular station( Overall) from our dataset.
#
df_station = data.groupby(['address'])['bluecar counter'].max().sort_values(ascending=False)

df_station.head(1)
```

```
# Finding the most popular station (At the most popular picking hour) from our dataset.
#
df_stati = data.groupby(['address'])['hour'].max().sort_values(ascending=False)

df_stati.head(1)
```

```
# Find which postal code is the most popular for picking up Blue cars? 
# Does the most popular station belong to that postal code? from our dataset.
#
df_station1 = data[['address', 'postal code', 'bluecar counter']]
df_postal = df_station1.sort_values('bluecar counter', ascending=False)
df_postal.head(1)
```

```
# Findb what charging status has the highest value for picking up Blue cars from our dataset.
#
df_city = data.groupby(['charging status'])['bluecar counter'].sum().sort_values(ascending=False)
df_city
```
