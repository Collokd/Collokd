```
# Importing the pandas library

import pandas as pd

# Importing Numpy library

import numpy as np
```

```
# Loading Telcom_Dataset

url = 'http://bit.ly/Telcom_dataset1'

df = pd.read_csv(url)

df.head(5)

df.rename(columns={"PRODUTC": "PRODUCT", "DATETIME":"DATE_TIME"}, inplace=True) # Renaming our columns as per the description

df.head(5)
```

```
# Loading Telcom_Dataset2

url2 = 'http://bit.ly/Telcom_dataset2'

df2 = pd.read_csv(url2, header=0)

df2.head(5)

df2.rename(columns={"DW_A_NUMBER": "DW_A_NUMBER_INT", "DW_B_NUMBER":"DW_B_NUMBER_INT"}, inplace=True) # Renaming our columns as per the description

df2.head(5)
```

```
# Loading Telcom_Dataset

url3 = 'http://bit.ly/Telcom_dataset3'

df3 = pd.read_csv(url3, header=0)

df3.head(5)

df3.rename(columns={"CELLID": "CELL_ID", "SIET_ID":"SITE_ID"}, inplace=True) # Renaming our columns as per the description

df3.head(5)

```

```
# Link the three datasets

frames = [df, df2, df3]
df_all = pd.concat(frames, axis=0, ignore_index=True)
df_all
```

```
# Dropping columns not needed 

df_all_new = df_all.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)

df_all_new
```

```
# Spliting Date and Time Column to date

df_all_new['DATE'] = df_all_new['DATE_TIME'].apply(lambda x:pd.Series([i for i in reversed (x.lower().split('"'))]))
df_all_new
df_all_new.drop(['DATE'], axis=1)
df_all_new['DATE'] = df_all_new['DATE_TIME'].str.split(" ", n=1, expand=True)
df_all_new
```

```
# Spliting Date and Time Column to TIME and drop DATE_TIME column

df_all_new['DATE'] = df_all_new['DATE_TIME'].str.split(" ", n=1, expand=True)
df_all_new['TIME'] = df_all_new['DATE_TIME'].str.split(" ").str[1]
df_all_new = df_all_new.drop(['DATE_TIME'], axis=1)
df_all_new
```

```
# Loading cells_geo dataset

df4 = pd.read_csv('cells_geo.csv', delimiter = ';') 
df4.head(5)
```

```
# Dropping columns not needed on cells_geo data set 

df4_new = df4.drop(['Unnamed: 0','STATUS', 'DECOUPZONE','LONGITUDE', 'LATITUDE', 'AREA', 'CELL_ID'], axis=1)

df4_new

df4_new.rename(columns={"VILLES": "CITY", "SITE_CODE": "SITE_ID"}, inplace=True) # Renaming COLUMN NAME VILLES TO CITY and SITE_CODE to SITE_ID

df4_new.head(5)
```

```
# Merging df_all_new(Telcom datasets) and df4_new(cells_geo)

df_merged = df_all_new.merge(df4_new, how='left', on='SITE_ID')

df_merged = df4_new.drop_duplicates(subset=['SITE_ID'], inplace=True)

df_merged = df_all_new.merge(df4_new, how='left', on='SITE_ID')

df_merged 

```

```
# Preview
df_merged.head(5)
```

```
# Identifying and dropping duplicates

d1 = df_merged.duplicated()

d1

d1 = df_merged.drop_duplicates()

d1
```

```
# Finding and deleting missing values

d2 = d1.isnull()

d2

df_no_missing = d2.dropna()
df_no_missing

df_cleaned = df_no_missing.dropna(how='all')
df_cleaned

df_cleaned.dropna(axis=1, how='all')

df_cleaned.fillna(0)
```
